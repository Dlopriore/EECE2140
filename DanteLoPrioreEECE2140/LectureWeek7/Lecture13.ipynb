```python
word = "joe mama boi dedicated"


str_list = word.split(" ")
out_str_list = list()
for item in str_list:
    if 'b' == item[0]:
        out_str_list.append(item)
    if 'ed' == item[-2:]:
        out_str_list.append(item)

print(out_str_list)

in_str = "This is a really big string word boned"
word_list = in_str.split(" ")

# One way 
wanted_words = list()
for cur_word in word_list:
    if cur_word.lower()[0] == 'b' or cur_word.lower()[-2:] == "ed":
        wanted_words.append(cur_word)
print(wanted_words)

# second way
wanted_words = list()
for cur_word in word_list:
    if cur_word.lower().startswith('b') or cur_word.lower().endswith("ed"):
        wanted_words.append(cur_word)
print(wanted_words)


# second way
wanted_words = list()
wanted_words = [(cur_word) for cur_word in word_list \
    if cur_word.lower().startswith('b') or cur_word.lower().endswith("ed")]

print(wanted_words)
print(' * '.join(wanted_words))


```

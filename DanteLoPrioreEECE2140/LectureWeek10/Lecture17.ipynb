```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 15, 400)

plt.plot([-1, 2, 5, 7, 9, 12], np.array([0, 10, 2, 25, 4, 5]), 'h--', c='hotpink', linewidth=3, markersize=10, mec='b')  #-,:,--,-
plt.plot(np.array([0, 10, 2, 25, 4, 5]), '<-.k') # is range(len): [0, 1, 2, ...]
plt.plot([-1, 2, 5, 7, 9, 12], np.array([0, 10, 2, 25, 4, 5]), '+:k',\
        x, x*np.sin(x), '-r', linewidth=2, markersize=10) # can plot more than one trendline
plt.grid()

plt.xlabel('x', fontsize=50, color='r')
plt.ylabel('y = f(x)')
plt.title('Figure 1', loc='center')

plt.grid(axis='both', color="darkgreen", linewidth=1, ls='dotted') # dashed, solid
#plt.grid(axis='both')
#plt.grid(axis='y = f(x)')
plt.axis([-1, 12, -20, 40])
plt.legend(['Points', 'y = x * sqrt(x)'])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_min = -2 * np.pi
x_max = 2 * np.pi

plt.figure()
x = np.arange(x_min, x_max, 0.001)
plt.plot(x, np.sin(x), 'r-', x, np.sin(2*x), 'k:', x, np.cos(x), '-b', x, np.cos(2*x), '-g')
plt.legend(["f(x) = sin(x)", "f(x) = sin(2x)", "f(x) = cos(x)", "f(x) = cos(2x)"])
plt.show()

x = np.arange(x_min, x_max, 0.01)
plt.subplot(321)
plt.plot(x, np.sin(x), "-r")
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of Sin(x)")

x = np.arange(x_min, x_max, 0.01)
plt.subplot(322)
plt.plot(x, np.cos(x), "-k")
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of Cos(x)")

x = np.arange(x_min, x_max, 0.01)
plt.subplot(323)
plt.plot(x, np.sin(2*x), "-g")
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of Sin(2x)")

x = np.arange(x_min, x_max, 0.01)
plt.subplot(324)
plt.plot(x, np.cos(2*x), "-b")
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of Cos(2x)")

plt.suptitle("Graphs of Sine and Cosine Functions")
```

```python
class Student:
    # Example of a Class

    # Properties or attributes
    # Usually class variables are for attributes and methods shared by all instances of the class.
    # objectJob = 'student'

    def __init__(self, fName, lName, studentId):
        # Constructor is used to assign values to object properties or object initialization.
        # Backbone of the class.

        # The self parameter is a reference to the current instance of the class.
        # It must be the first parameter of any function/method in a class
        # Instance variables are for data unique to each instance

        self.firstName = fName
        self.lastName = lName
        self.name = (fName, lName)
        self.studentId = studentId
        self.grades = []
        self.gpa = 0

    def update_grades(self, *grades):
        for cur_grade in grades:
            self.grades.append(cur_grade)
    

std1 = Student("Dante", "LoPriore", 1234567)
print(f'First Name: {std1.firstName} | Last Name: {std1.lastName} | Student Id: {std1.studentId}')
print(std1.name)
print(f'Here is how an object of this class is printed: {std1}')
print(f'Here is how an object of this class is printed: {std1.__init__}')


print(f'Before Update {std1.grades}')
std1.update_grades(3, 3.6)
print(f'After Update {std1.grades}')


```

```python
class Car:
    def __init__(self, name, model, color):
        self.name = name
        self.model = model
        self.mileage = 0
        self.color = color

my_car1 = Car("Toyota", 2016, "Red")

print(f'{my_car1.name} | {my_car1.model} | {my_car1.color} | {my_car1.mileage}')

my_car1.model = 2023
my_car1.color = 'Black'
my_car1.mileage = 60000
print(f'{my_car1.name} | {my_car1.model} | {my_car1.color} | {my_car1.mileage}')
```

```python
import numpy as np

class Polygon:

    def __init__(self, num_sides):
        self.num_sides = num_sides
        self.sides = []
        self.perimeter = 0

    def __str__(self):
        return f'The polygon has {self.num_sides} sides.'
    
    def setSides(self):
        for individual_side in range(self.num_sides):
            cur_side = float(input(f'Enter the length of the {individual_side+1}th side (+Number): '))
            self.sides.append(cur_side)
    
    def find_parameter(self):
        self.perimeter = sum(self.sides)
        
class Triangle(Polygon):

    def __init__(self):
        Polygon.__init__(self, num_sides = 3)
        super().setSides()
        self.find_area()

    
    def __str__(self):
        return f'The triangle has an area of {self.area}.'
    
    def find_area(self):
        s = sum(self.sides)/2
        self.area = np.sqrt(s * (s-self.sides[0]) * (s-self.sides[1]) * (s-self.sides[2]))

tri1 = Triangle()
tri1.setSides()
tri1.find_parameter()
print(tri1)
print(tri1.perimeter)


```
